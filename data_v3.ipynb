```python
from urllib.request import urlopen
from urllib.error import HTTPError, URLError
from urllib.parse import urlparse
from bs4 import BeautifulSoup
from urllib.parse import unquote
```

```python
urlList = []
pg = 1
while(True):
    print(pg)
    if pg == 5: break
    html = urlopen('https://www.jobtopgun.com/AjaxServ?pg=' + str(pg))
    bsObj = BeautifulSoup(html.read(),"lxml")
    links = bsObj.find_all('div', {'class':'jobListPositionName'})
    if len(links) == 0:
        break
    for link in links:
        path = unquote(link.find('a')['href']).replace(' ',"%20")
        url = 'https://www.jobtopgun.com/' + path
        urlList.append(url)
    pg += 1
```

```python
posList = [] #keep html
urlList = [] #keep url
jobList = [] # keep position name
firmList = [] #keep firm name
salList = [] # keep salary
eduList = [] # keep edu
locList = [] #keep location

def striptext(x):
    u = x.get_text().replace('\t','').strip()
    return u
pg = 1
while True:
    if pg == 5: break
    html = urlopen('https://www.jobtopgun.com/AjaxServ?pg='+str(pg))
    bsObj = BeautifulSoup(html, "lxml")
    check = bsObj.find_all('div', {'class':'jobListPositionName'})
    if len(check) == 0:
        break
    posList += bsObj.findAll('div',{'class':'jobListPositionName'})
    for posName in check:
        path = unquote(posName.find('a')['href']).replace(' ',"%20")
        url = 'https://www.jobtopgun.com' + path
        urlList.append(url)
        
        firmName = posName.find_next('div')
        salName = firmName.find_next('div')
        eduName = salName.find_next('div')
        locName = eduName.find_next('div')
        
        pos = posName.find('a').get_text().strip()
        print(pos)
        jobList.append(pos)
        firmList.append(striptext(firmName))
        salList.append(striptext(salName))
        eduList.append(striptext(eduName))
        locList.append(striptext(locName))
    pg += 1  
```
